```python
import time
kelimeler = [
    "kitaplık", "saksı", "orkide", "parfüm", "telefon", "buzdolabı", "ampul",
    "mirket", "asansör", "kumandan", "operatör"
    "merdane", "değirmen", "pencere", "portakal", "çekirdek", "etajer",
    "kafeterya", "lokanta", "karnabahar", "brokoli", "timsah", "fenerbahçe",
    "sarımsak", "kanarya", "tülbent", "papatya", "okyanus", "hastane",
    "biyoloji", "antoloji", "hapşuruk", "öksürük", ""
]

def oyun_hazirlik():
    global secilen_kelime, gorunen_kelime, can
    import random
    secilen_kelime = random.choice(kelimeler)
    gorunen_kelime = ["-"] * len(secilen_kelime)
    can = 5

def harf_al():
    """Kullanıcıdan bir harf alır, alana kadar gerekirse hata verir, birisi quit yazarsa programı kapatır"""
    devam = True
    while devam:

        harf = input("Bir harf giriniz: ")
        if len(harf) == 1 and harf.isalpha() and harf not in gorunen_kelime:
            devam = False
        else:
            print("Hatalı Giriş")
    return harf.lower()

def oyun_dongusu():
    """Alınan harf kelimede varsa görünen karakterler listesi güncellenir,
     yoksa can azaltılır, ve bu döngü can bitene kadar ya da kelime bilinene kadar devam eder..."""
    global gorunen_kelime, can
    while can > 0 and secilen_kelime != "".join(gorunen_kelime):
        print("")
        print(
            "kelime: " + "".join(gorunen_kelime))

        girilen_harf = harf_al()
        pozisyonlar = harf_kontrol(girilen_harf)
        if pozisyonlar:
            for p in pozisyonlar:
                gorunen_kelime[p] = girilen_harf
        else:
            can -= 1
            print("Yanlış tahmin! Kalan tahmin hakkın: " + str(can))
            switch(can)

def harf_kontrol(girilen_harf):
    """Gelen harfin seçilen kelimede nerelerde olduğunu bulur"""
    poz = []
    for index, h in enumerate(secilen_kelime):
        if h == girilen_harf:
            poz.append(index)
    return poz

def main():
    tekrar = True
    print("Merhaba, Adam Asmaca oyununa hoşgeldiniz.")
    print("")
    name = input("İsminiz nedir? ")
    print("")
    print("Hoşgeldiniz " + name)
    print("")
    time.sleep(1)
    print("Haydi Başlayalım!")
    time.sleep(1)
    while tekrar:
        oyun_hazirlik()
        oyun_dongusu()
        oyun_sonucu()
        if input("Devam?(e/h) ").lower() == "h":
            tekrar = False
            print("Teşekkürler. Yeniden görüşmek dileğiyle :)")

def oyun_sonucu():

    if can > 0:
        print(secilen_kelime)
        print("Kazandınız")
    else:
        print("Kaybettiniz")

def switch(can):
    if can == 5:
        time.sleep(1)
        print("   _____ \n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "__|__\n")
    elif can == 4:
        time.sleep(1)
        print("   _____ \n"
              "  |     |\n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "__|__\n")
    elif can == 3:
        time.sleep(1)
        print("   _____ \n"
              "  |     | \n"
              "  |     |\n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "__|__\n")
    elif can == 2:
        time.sleep(1)
        print("   _____ \n"
              "  |     | \n"
              "  |     |\n"
              "  |     | \n"
              "  |      \n"
              "  |      \n"
              "  |      \n"
              "__|__\n")
    elif can == 1:
        time.sleep(1)
        print("   _____ \n"
              "  |     | \n"
              "  |     |\n"
              "  |     | \n"
              "  |     O \n"
              "  |      \n"
              "  |      \n"
              "__|__\n")
    elif can == 0:
        time.sleep(1)
        print("   _____ \n"
              "  |     | \n"
              "  |     |\n"
              "  |     | \n"
              "  |     O \n"
              "  |    /|\ \n"
              "  |    / \ \n"
              "__|__\n")
main()
```
